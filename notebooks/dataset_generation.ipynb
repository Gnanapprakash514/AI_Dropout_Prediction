```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 500

data = []

for _ in range(n):
    attendance = np.random.randint(40, 61)
    prev_marks = np.random.randint(40, 61)
    marks = np.clip(prev_marks + np.random.randint(-10, 10), 40, 60)

    assignment = np.clip(attendance + np.random.randint(-10, 10), 40, 60)

    study_hours = np.random.randint(1, 6)
    backlogs = np.random.randint(0, 5)
    mental_stress = np.random.randint(1, 11)

    performance_drop = prev_marks - marks

    risk = 0
    if attendance < 50:
        risk += 2
    if marks < 50:
        risk += 2
    if performance_drop > 5:
        risk += 2
    if backlogs >= 3:
        risk += 2
    if study_hours <= 2:
        risk += 1
    if mental_stress >= 7:
        risk += 2
    if assignment < 50:
        risk += 1

    dropout = 1 if risk >= 5 else 0

    data.append([
        attendance, marks, prev_marks,
        backlogs, mental_stress,
        study_hours, assignment, dropout
    ])

df = pd.DataFrame(data, columns=[
    "attendance", "marks", "prev_marks",
    "backlogs", "mental_stress",
    "study_hours", "assignment", "dropout"
])

df.head()
```

```python
df.to_csv("../data/raw/student_data.csv", index=False)
```
